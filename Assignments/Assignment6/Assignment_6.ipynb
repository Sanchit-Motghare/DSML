```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/temperaturess.csv')
df
```

```python
df.describe()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
x = df["YEAR"].values.reshape(-1, 1)
y = df["JAN"].values.reshape(-1,1)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2,random_state=0)
regressor = LinearRegression()
regressor.fit(x_train, y_train)
y_pred = regressor.predict(x_test)
print(regressor.intercept_)
print(regressor.coef_)
plt.scatter(x_test, y_test)
plt.plot(x_test,y_pred,color="green")
plt.xlabel('Temperature')
plt.ylabel('Year')
plt.title('Linear Regression Model For January')
plt.show()
print("Mean Absolute Error: ", metrics.mean_absolute_error(y_test, y_pred))
print("Mean Squared Error: ", metrics.mean_squared_error(y_test, y_pred))
print("R2 Score: ", metrics.r2_score(y_test, y_pred))
```

```python
months = ['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC']
all_metrics = []
for month in months:
    x = df["YEAR"].values.reshape(-1, 1)
    y = df[month].values.reshape(-1, 1)
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
    regressor = LinearRegression()
    regressor.fit(x_train, y_train)
    y_pred = regressor.predict(x_test)
    plt.scatter(x_test, y_test)
    plt.plot(x_test, y_pred, color="green")
    plt.xlabel('Year')
    plt.ylabel('Temperature')
    plt.title(f'Linear Regression Model For {month}')
    plt.show()
    metrics_dict = {
        'Month': month,
        'Mean Absolute Error': metrics.mean_absolute_error(y_test, y_pred),
        'Mean Squared Error': metrics.mean_squared_error(y_test, y_pred),
        'R2 Score': metrics.r2_score(y_test, y_pred)}
    all_metrics.append(metrics_dict)
for metrics_dict in all_metrics:
    print(f"\nMetrics for {metrics_dict['Month']}:\n")
    print("Mean Absolute Error:", metrics_dict['Mean Absolute Error'])
    print("Mean Squared Error:", metrics_dict['Mean Squared Error'])
    print("R2 Score:", metrics_dict['R2 Score'])
```
