```python
import pandas as pd
import matplotlib
import scipy
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('car_crashes')
print(df.head())
```

```python
sns.relplot(data=df,x="alcohol",y="speeding")
sns.set_style("whitegrid")
plt.show
```

```python
sns.boxplot(y='ins_premium',data=df,width=0.5)
```

```python
sns.barplot(x='abbrev',y='speeding',data=df,width=0.5)
```

```python
sns.histplot(data=df,x="ins_losses",kde=True)
```

```python
sns.lineplot(df,x="alcohol",y="speeding")
```

```python
sns.pairplot(df)
```
